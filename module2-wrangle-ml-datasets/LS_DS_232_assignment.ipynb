```python
# The Pandas Profiling Report showed that SALE_PRICE was read as strings
# Convert it to integers
df_sub['SALE_PRICE'] = df_sub['SALE_PRICE'].str.replace("-","")
df_sub['SALE_PRICE'] = df_sub['SALE_PRICE'].str.replace("$","")
df_sub['SALE_PRICE'] = df_sub['SALE_PRICE'].str.replace(",","",regex=True)
df_sub['SALE_PRICE'] = df_sub['SALE_PRICE'].astype(int)
```
